# TMDB movies

Load all the necessary packages

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Load the dataset into a DataFrame and inspect the first rows of the DataFrame.

```python
movies = pd.read_csv('TMDB-movies-small.csv', encoding='ISO-8859-1')
movies.head()
```

Create a new DataFrame with all the movies released in 2000. Tip: Format the release dates as pandas datetime objects. Inspect the amount of rows in the new DataFrame.

```python
# Format column to datetime 
movies['release_date'] = pd.to_datetime(movies['release_date'])

# Extract year
movies['year'] = movies['release_date'].dt.year

# Filter by year
movies_2000 = movies[movies['year'] == 2000]
movies_2000.head()
```

Create a new DataFrame df_topmovies which only incorporates the rows with the movies of the year 2000 who get an average score of at least 9. Compare the total rows of the original dataframe with df_topmovies. (Tip: use shape)

```python
# Filter by condition
df_topmovies = movies_2000[movies_2000['vote_average'] >= 9]

# Compare total rows
print("Total rows of dt_topmovies:")
print(df_topmovies.shape[0])

print("Total rows of original dataframe:")
print(movies.shape[0])
```

Replace all NaN's with 0.

```python
# Handling missing data
clean_movies = movies.fillna(0) 
# or movies.replace(np.nan, 0)

# Check
clean_movies.isnull().sum()
```

Plot the data using a graph with on the x-axis the release dates and on the y-axis the runtime of the film.
* Use appropriate labels and titles
* Change the ticks and tick labels so every month has a tick and as label the abbreviation of the month (Jan, Feb, March etc.)
* Change the color of the graph
* Show a grid
* Show the legend

```python
# Extract the month from the release_date and create a new column 'month'
clean_movies['month'] = clean_movies['release_date'].dt.month

# Calculate average runtime for each month
average_runtime = clean_movies.groupby('month')['runtime'].mean()

# Prepare
X = range(1, 13) # for the 12 months
Y = average_runtime

# Plot/Render
fig, ax = plt.subplots()
ax.bar(X, Y, label='Runtime', color='royalblue', width=0.6)

# Set x-axis ticks and labels
ax.set_xticks(range(1, 13))
ax.set_xticklabels(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])

# Set y-axis to start at 50
ax.set_ylim(50)

# Set labels and title
ax.set_xlabel('Release Date')
ax.set_ylabel('Average Runtime')
ax.set_title('Average Runtime of Films Over Time')

# Show grid and legend
ax.grid(True)
ax.legend()

plt.show()
```
